```python
import pandas as pd
import os
from typing import List, Optional
```

```python
#Specify Excel file path
attendance_v3_file_path = "../../data/Attendance_Template_v3.xlsx"
#Load in Excel file
attendance_v3_file = pd.ExcelFile(attendance_v3_file_path)
#Iterate through each sheet in Excel file
for sheet_name in attendance_v3_file.sheet_names:
    #Read in the current sheet
    mentor_df = pd.read_excel(attendance_v3_file_path, sheet_name=sheet_name)
    #Using the sheet name, define the CSV output file
    csv_output_file = f"../../data/{sheet_name}.csv"
    #Save the df to CSV
    mentor_df.to_csv(csv_output_file, index=False)
```
